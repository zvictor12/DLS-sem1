# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer


pip_num = Pipeline([
    ("impute", IterativeImputer(n_nearest_features=2, imputation_order="ascending", random_state=42)),
    ("scaler", StandardScaler())
])

pip_lab = Pipeline([
    ("encoder", OrdinalEncoder(categories="auto",
    handle_unknown="use_encoded_value",
    unknown_value=-1)),
    ("scaler", StandardScaler())
])

pip_ohe = Pipeline([
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

transf = ColumnTransformer([
    ("num", pip_num, num_cols),
    ("lab", pip_lab, lab_cols),
    ("ohe", pip_ohe, ohe_cols)
],
remainder="passthrough")
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = 'iframe' #это для каггла, в гугл диске мб будет мешать

reg_list = np.linspace(0, 1, 5)
model = LogisticRegressionCV(penalty="elasticnet", scoring="roc_auc", solver="saga", max_iter=500, n_jobs=-1, random_state=42, l1_ratios=reg_list)
model.fit(X_train, y_train)

scores = np.mean(model.scores_[1], axis=0)
Cs = model.Cs_
sort_ind = np.argsort(Cs)
Cs = Cs[sort_ind]
scores = scores[sort_ind]

fig = go.Figure(data=[go.Surface(z=scores, x=Cs, y=reg_list)])
fig.update_layout(
    title="Средний (по батчам) ROC-AUC по log C и l1_ratio",
    scene=dict(
        xaxis_title="log C",
        yaxis_title="l1_ratio",
        zaxis_title="ROC-AUC",
        xaxis=dict(type='log')
    )
)
fig.show()
```

```python
i, j = np.unravel_index(np.argmax(scores), scores.shape)
print("Best C: %.1f" % Cs[i])
print("Best l1_ratios: %.1f" % reg_list[j])
if reg_list[j] < 0.5:
    print("L2 is better")
else:
    print("L1 is better")
print("Best ROC-AUC: %.3f" % scores[i, j])
```
